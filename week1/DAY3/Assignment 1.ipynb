###CASE AND WHEN CHAINING PROBLEMS

```sql
---1
select *,
case 
when department='Finance' then
  case when performance_rating='A' then salary*1.2
       when performance_rating='B' then salary*1.15
       when performance_rating='C' then salary*0.05 end
when department='Engineering' then 
  case when performance_rating='A' then salary*1.18
       when performance_rating='B' then salary*1.12
       when performance_rating='C' then salary*0.03 end
else salary*1.1 end as bonus_salary
from employee
```

###Nested case and when


```sql
---1
select *,
case
when performance_rating='A' then
  case when salary>80000 and experience>9 then salary*1.25
       when salary between 50000 and 80000 then salary*1.15
       else salary*1.2 end
when performance_rating='B' then
  case when experience>5 then salary*1.12
       else salary*1.1 end
else salary*1.1 end as salary_hike
from employee
```

```sql
---2
select *,
case 
when department='Finance' then
  case when performance_rating='A' and experience>10 then salary*1.25
       else salary*1.2 end
when department='HR' then 
  case when performance_rating='B'and experience>5 then salary*1.15
      else salary*1.1 end
else salary*0.08 end as bonus_salary
from employee
```

```sql
---3
SELECT emp_name, salary, experience, performance_rating,
CASE 
    WHEN salary > 70000 THEN
        CASE 
            WHEN performance_rating = 'A' AND experience > 8 THEN 'Top Performer'
            WHEN experience < 8 THEN 'Mid Performer'
            ELSE 'High Performer'
        END
    WHEN salary BETWEEN 50000 AND 70000 THEN
        CASE 
            WHEN performance_rating = 'A' THEN 'Rising Star'
            ELSE 'Average Performer'
        END
    ELSE 'Low Performer'
END AS employee_category
FROM Employee;
```

```sql
---4
SELECT emp_name, salary, experience,
CASE 
    WHEN salary > 90000 THEN
        CASE 
            WHEN experience > 10 THEN '35% Tax'
            ELSE '30% Tax'
        END
    WHEN salary BETWEEN 60000 AND 90000 THEN
        CASE 
            WHEN experience > 5 THEN '25% Tax'
            ELSE '20% Tax'
        END
    ELSE '15% Tax'
END AS tax_bracket
FROM Employee;
```

```sql
---5
SELECT emp_name, salary, experience, performance_rating,
CASE 
    WHEN performance_rating = 'A' THEN
        CASE 
            WHEN salary > 75000 AND experience > 7 THEN 'Eligible for Senior Role'
            ELSE 'Eligible for Junior Role'
        END
    WHEN performance_rating = 'B' THEN
        CASE 
            WHEN experience > 5 THEN 'Eligible for Consideration'
            ELSE 'Not Eligible'
        END
    ELSE 'Not Eligible'
END AS promotion_status
FROM Employee;
```
